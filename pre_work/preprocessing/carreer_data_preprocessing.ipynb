```python
#resultado tabla de carreras(id,name_descr(300),id_university,arancel,num_semestres,
                            #area,horario_bin,empleabilidad,ingreso,corte,coordenadas_campus)
#usar gower distance para el arbol.
import numpy as np
import pandas as pd
import os
```

```python
carreer_vectors = pd.read_csv('../preprocessed_data/vector_carreers.csv')
carreers = pd.read_csv('../preprocessed_data/carreras2.csv')
campus = pd.read_csv('../data/tuni/campuses.tsv', sep='\t')
```

```python
carreer_data = carreers[['ID','ID Universidad','ID Campus','Cant. Semestres','Arancel','Empleabilidad','Ingreso','Corte 2017','area_id']]
campus = campus[['ID','Latitud','Longitud']]
```

```python
#carreer_data = carreer_data.join(carreer_vectors.set_index('id'),on='ID')
carreer_data = carreer_data.join(campus.set_index('ID'),on='ID', rsuffix='_campus',lsuffix='_carreer')
carreer_data['Empleabilidad'] = carreer_data.apply(lambda x: float(x['Empleabilidad'].replace(',','.')) if type(x['Empleabilidad']) == str else x['Empleabilidad'],axis=1)
carreer_data['Corte 2017'] = carreer_data.apply(lambda x: float(x['Corte 2017'].replace(',','.')) if type(x['Corte 2017']) == str else x['Corte 2017'],axis=1)
carreer_data['Latitud'] = carreer_data.apply(lambda x: float(x['Latitud']) if type(x['Latitud']) == str else x['Latitud'],axis=1)
carreer_data['Longitud'] = carreer_data.apply(lambda x: float(x['Longitud']) if type(x['Longitud']) == str else x['Longitud'],axis=1)
carreer_data['Arancel'] = carreer_data.apply(lambda x: float(x['Arancel'].replace('.','')) if type(x['Arancel']) == str else (x['Arancel']),axis=1)
```

```python
carreer_data = carreer_data.fillna(carreer_data.mean())
```

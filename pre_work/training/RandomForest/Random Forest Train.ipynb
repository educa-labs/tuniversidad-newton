```python
import numpy as np
import pandas as pd
import os
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier 
from sklearn.preprocessing import scale,LabelEncoder
from sklearn.model_selection import StratifiedShuffleSplit,train_test_split
from sklearn.metrics import accuracy_score
from sklearn.externals import joblib
```

```python
def print_metrics(y_true,y_pred):
    print("accurracy: {}".format(accuracy_score(y_true,y_pred)))
    #print("f1: {}".format(f1_score(y_true,y_pred,average="macro")))   
```

```python
data_folder = '../../preprocessed_data/area_data'
result_folder = '../../serialized/forest'
```

```python
test_scores = np.array(pd.read_csv('test_recomendations.csv'))
for file in os.listdir(data_folder):
    data = pd.read_csv(os.path.join(data_folder, file))
    X = np.array(data[["mat","len","cie","his","nem"]])
    #y = np.array(data.apply(lambda x: "{}/{}".format(x['universidad'],x['tuni_carreer_name']),axis=1))
    y = np.array(data['tuni_carreer_id'])
#     lb = LabelEncoder()
#     lb.fit(y)
#     y_encoded = lb.transform(y)
    #X_scaled = scale(X)
    #randomized stratified test/train split
    print(file)
    sss = StratifiedShuffleSplit(n_splits=5, train_size=0.8, random_state=0)
    try:
        for train_index,test_index in sss.split(X, y):
            X_train, X_test = X[train_index], X[test_index]
            y_train, y_test = y[train_index], y[test_index]
            model = RandomForestClassifier(25,min_samples_split=5,n_jobs=-1)
            model.fit(X_train,y_train)
            result = model.predict(X_test)
            print_metrics(y_test,result)
    except ValueError:
        print("{} failed!".format(file))
    model = RandomForestClassifier(25,min_samples_split=5,n_jobs=-1)
    model.fit(X,y)
    clases = model.classes_
    result = model.predict_proba(test_scores)
    for i in range(test_scores.shape[0]):
        print(test_scores[i])
        best = np.argpartition(result[i],-3)[-3:]
        for index in best:
            print("{}, prob: {}".format(clases[index],result[i][index]))
    joblib.dump(model,os.path.join(result_folder,os.path.splitext(file)[0]))
    
```
